```python
class Bebida:
  def __init__(self, nombre, marca, mililitros):
    self.nombre = nombre
    self.marca = marca
    self.mililitros = mililitros

  def leerdatos(self):
    print("Por favor, ingrese los datos de la bebida:")
    self.nombre = input("Ingrese el nombre de la bebida: ")
    self.marca = input("Ingrese la marca de la bebida: ")
    self.mililitros = input("Ingrese los mililitros de la bebida: ")

  def mostrardatos(self):
    print("Datos de la Bebida:")
    print(f"Nombre: {self.nombre}")
    print(f"Marca: {self.marca}")
    print(f"Mililitros: {self.mililitros}")

# Fin de la clase Bebida

# Creando los objetos bebida1, bebida2

# Inicializamos los objetos
bebida1 = Bebida("", "", "")
bebida2 = Bebida("", "", "")

print("Bebida Numero 1")

bebida1.leerdatos()
bebida1.mostrardatos()
print("-------------------------------------------------")

print("Bebida Numero 2")
bebida2.leerdatos()
bebida2.mostrardatos()
print("-------------------------------------------------")

print("programa realizado por Cristopher Lopez NL = 33")
```
